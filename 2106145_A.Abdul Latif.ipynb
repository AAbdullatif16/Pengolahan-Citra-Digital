```
from PIL import Image

def tampilkan_citra(r, g, b, M, N):
    """Menampilkan citra yang berukuran M x N pixel ke layar."""
    width = N
    height = M
    image = Image.new("RGB", (width, height))

    for i in range(M):
        for j in range(N):
            set_pixel(image, j, i, r[i][j], g[i][j], b[i][j])

    image.show()

def set_pixel(image, i, j, r, g, b):
    """Menampilkan pixel dengan komponen rgb pada koordinat i, j."""
    image.putpixel((i, j), (r, g, b))

# Contoh penggunaan:
M = 480
N = 640

# Membuat citra R, G, B
r = [[255 for _ in range(N)] for _ in range(M)]
g = [[0 for _ in range(N)] for _ in range(M)]
b = [[0 for _ in range(N)] for _ in range(M)]

# Menampilkan citra
tampilkan_citra(r, g, b, M, N)
```

```
def baca_citra_dari_arsip(nama_arsip):
    """Membaca citra dari file citra mentah. Citra hasil pembacaan disimpan di dalam matriks f."""
    try:
        with open(nama_arsip, "rb") as fp:
            M = int.from_bytes(fp.read(2), byteorder='little', signed=False)  # Baca tinggi citra
            N = int.from_bytes(fp.read(2), byteorder='little', signed=False)  # Baca lebar citra

            f = alokasi(M, N)  # Alokasi memori matriks untuk citra f

            if f is None:
                print("Memori tidak cukup")
                return None

            # Baca data citra baris demi baris
            for i in range(M):
                # Baca data citra baris ke-i
                f[i] = list(fp.read(N))

            return f

    except FileNotFoundError:
        print(f"Arsip '{nama_arsip}' tidak ditemukan")
        return None

def alokasi(M, N):
    """Mengalokasikan memori untuk matriks M x N."""
    try:
        return [[0 for _ in range(N)] for _ in range(M)]
    except MemoryError:
        return None

nama_arsip = "nama_file.dat"  # Ganti dengan nama file dan ekstensi Anda
citra = baca_citra_dari_arsip(nama_arsip)

if citra is not None:
    # Lanjutkan dengan pemrosesan citra jika berhasil dibaca
    print("Citra berhasil dibaca.")
else:
    print("Gagal membaca citra.")
```

Dari keseluruhan kode ini fungsinya untuk membaca citra dari sebuah file citra mentah dan menyimpannya dalam matriks citra.

```
def tulis_citra_ke_arsip(nama_arsip, f):
    """Menulis citra f ke dalam arsip nama_arsip."""
    try:
        M = len(f)
        N = len(f[0])

        with open(nama_arsip, "wb") as fp:
            fp.write(M.to_bytes(2, byteorder='little'))  # Tulis tinggi citra
            fp.write(N.to_bytes(2, byteorder='little'))  # Tulis lebar citra

            # Tulis data citra baris demi baris
            for i in range(M):
                # Tulis data citra baris ke-i
                fp.write(bytes(f[i]))

    except Exception as e:
        print(f"Terjadi kesalahan saat menulis arsip: {e}")

# Contoh penggunaan:
nama_arsip = "nama_file.dat"  # Ganti dengan nama file dan ekstensi Anda
citra = [[255, 0, 0], [0, 255, 0], [0, 0, 255]]  # Contoh citra
tulis_citra_ke_arsip(nama_arsip, citra)
```
